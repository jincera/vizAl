#### PREGUNTA 7. Muestre el código necesario para obtener el siguiente gráfico:

```r
x<- seq(-pi,pi,0.1)
fn_cos<-cos(x)
fn_sin<-sin(x)
fn_cos_2<-cos(x)/2
plot(main= 'Gráficas superpuestas',ylab='', xlab='X en [-Pi, Pi]',x,fn_sin,col='red',type='l')
legend("topleft", legend=c('sin(x)','cos(x)','cos(x)/2'),fill=c('red','black','blue'))
lines(x,fn_cos,col='black',type='l')
lines(x,fn_cos_2,col='blue',type='l')
```

#### PREGUNTA 8. Añada al gráfico anterior dos líneas verticales verdes cortadas donde cos(x) y sin(x) se intersectan

```r
x<- seq(-pi,pi,0.1)
fn_cos<-cos(x)
fn_sin<-sin(x)
fn_cos_2<-cos(x)/2
plot(main= 'Gráficas superpuestas',ylab='', xlab='X en [-Pi, Pi]',x,fn_sin,col='red',type='l')
legend("topleft", legend=c('sin(x)','cos(x)','cos(x)/2'),fill=c('red','black','blue'))
lines(x,fn_cos,col='black',type='l')
abline(v=pi/4,col='green')
abline(v=-3*pi/4,col='green')
lines(x,fn_cos_2,col='blue',type='l')
```
